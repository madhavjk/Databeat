```python
import sqlalchemy as db

#Establishing connection with the database.
engine = db.create_engine('mysql://root@localhost/organization')
connection = engine.connect()
metadata = db.MetaData()
```

```python
#referencing employee tables to "employee".
employee = db.Table('employee', metadata, autoload=True, autoload_with=engine)
```

```python
import pandas as pd
# Fetching and printing all records
query = db.select([employee]) 
ResultProxy = connection.execute(query)
Result = ResultProxy.fetchall()
df = pd.DataFrame(Result, columns = Result[0].keys())
```

```python
df
```

```python
# Fetching employee names having basic salary more than 5000

query = db.select([employee.columns.e_name, employee.columns.basic_salary]).where(employee.columns.basic_salary > 5000)
resultproxy2 = connection.execute(query)
result2 = resultproxy2.fetchall()
df2 = pd.DataFrame(result2, columns = result2[0].keys())
df2
```

```python
# Fetching distinct names from the employee table
query = db.select([employee.columns.e_name.distinct()])
resultproxy3 = connection.execute(query)
result3 = resultproxy3.fetchall()
df3 = pd.DataFrame(result3, columns = ["Name"])
df3
```

```python
# Fetching Employees whose basic_salary equals to 2500 or 6500 or 10000
query = db.select([employee.columns.e_name, employee.columns.basic_salary]).where(employee.columns.basic_salary.in_([2500, 6500, 10000]))
resultproxy4 = connection.execute(query)
result4 = resultproxy4.fetchall()
df4 = pd.DataFrame(result4, columns = result4[0].keys())
df4
```

```python
# Fetching number of employees working in each department
query = db.select([employee.columns.department, db.func.count(employee.columns.e_name).label("count")]).group_by(employee.columns.department)
resultproxy5 = connection.execute(query)
result5 = resultproxy5.fetchall()
df5 = pd.DataFrame(result5, columns = result5[0].keys())
df5
```

```python
# Arranging all the employees in ascending order of their basic salaries.
query = db.select([employee]).order_by(employee.columns.basic_salary)
resultproxy6 = connection.execute(query)
result6 = resultproxy6.fetchall()
df6 = pd.DataFrame(result6, columns = result6[0].keys())
df6
```
